# Pandemic Dynamics

## Introduction to 'nonlinear' dynamics

* Logistic Growth equation (Verhulst Equation)
* Kermack McKendrick 
* Imperfect Immunity (immunological forgetting)

* Nullclines

```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline
pylab.rcParams['figure.figsize'] = (12,10)
```

```python
#Create a function for log equation
def verhulst (state, t):
    I = state[0] #the number of infected individuals
    dIdt = a * I * (N-I) #speed at which virus spreads
    return(dIdt)
```

```python
#pick parameters
a = 0.4
N = 100 #population size
state0 = 1
```

```python
#Run simulations
from scipy.integrate import odeint

t = np.linspace(0, 0.3, 10000)
state = odeint(verhulst, state0, t)

#plot results
plt.plot(t, state, linewidth = 5)
plt.plot(t, N*np.ones(t.shape), linestyle = ':', linewidth = 5)
```

# Kermack Mckendrick 

### We add a term for recovery (immunity) after infection

$\dot{I} = a I S - \frac{I}{tau}$

$\dot{S} = -a I S$

```python
def kermack(state,t):
    I = state[0]
    S = state[1]
    
    #define system
    dIdt = a*I*S - I/tau
    dSdt = -a*I*S
    
    return(dIdt, dSdt)
```

```python
a = 0.5
tau = 0.1

#We need to initial conditions
state0 = [1,99] #1 infected 99 susceptible

t = np.linspace(0,1,1000)

state = odeint(kermack, state0, t)

N = np.sum(state0)
def Rt(N, It, St):
    Nvec = N*np.ones(St.shape)
    return(Nvec - St - It)

plt.plot(t, state, linewidth = 5)
plt.plot(t, Rt(N,state[:,0], state[:,1]), linewidth =5)
plt.legend(('Infected', 'Susceptible', 'Recoevered'))
```

```python
#Infected is the curve to flatten (when people say flatten the curve)

a = 0.1 #changing 'a' is the only way to change the curve since I and S will be constant numbers
tau = 0.1

#We need to initial conditions
state0 = [1,99] #1 infected 99 susceptible

t = np.linspace(0,1,1000)

state = odeint(kermack, state0, t)

N = np.sum(state0)
def Rt(N, It, St):
    Nvec = N*np.ones(St.shape)
    return(Nvec - St - It)

plt.plot(t, state, linewidth = 5)
plt.plot(t, Rt(N,state[:,0], state[:,1]), linewidth =5)
plt.legend(('Infected', 'Susceptible', 'Recoevered'))

#When infection velocity > than recovery rate thats mathematical definition of outbreak
print('Infection velocity = ', a*N) 
print('Recovery velocity = ', 1/tau)
```

```python
#plot phase diagram
a = 0.5
Ilim = [0,100]
Slim = [0,100]

h1 = np.linspace(Ilim[0], Ilim[1], 20)
h2 = np.linspace(Slim[0], Slim[1], 20)

H1, H2 = np.meshgrid(h1,h2)

u,v = np.zeros(H1.shape), np.zeros(H2.shape)

NJ, NK = H1.shape

for j in range(NJ):
    for k in range(NK):
        Istate = H1[j,k]
        Sstate = H2[j,k]
        
        statejk = [Istate, Sstate]
        
        ISdot = kermack(statejk, [])
        u[j,k] = ISdot[0]
        v[j,k] = ISdot[1]

#Normalization
M = np.hypot(u,v)
M[M == 0] == 1.
u /= M
v /= M

plt.quiver(H1, H2, u,v, color = 'red')
plt.xlabel('Infected individuals', fontsize = 15)
plt.ylabel('Susceptible individuals', fontsize = 15)
plt.show()
```

```python
state = odeint(kermack, state0, t)

plt.plot(state[:,0], state[:,1], linewidth = 5)

plt.quiver(H1, H2, u,v, color = 'red')
plt.xlabel('Infected individuals', fontsize = 15)
plt.ylabel('Susceptible individuals', fontsize = 15)
plt.show()
```

```python
Inullcline = 1/(tau*a) #Peak level of infections!

plt.plot(state[:,0], state[:,1], linewidth = 5)
plt.plot(np.linspace(0,100,100), Inullcline*np.ones(100), linewidth = 5) #Plot of peak infections 

plt.quiver(H1, H2, u,v, color = 'red')
plt.xlabel('Infected individuals', fontsize = 15)
plt.ylabel('Susceptible individuals', fontsize = 15)
plt.show()
```

## Imperfect Immunity

### Pool of immune memmory cells degrates over time

```python
def imperfectImmunity(state,t):
    I = state[0]
    S = state[1]
    
    dIdt = a*I*S - I/tau
    R = N - S - I # Recoverd population at time t
    dSdt = -a*I*S + R/tau2 #immunological forgetting term
    
    return(dIdt, dSdt)
```

```python
a = 0.4
tau = 0.1
tau2 = 3 #30x rate of the immunological loss when compared to memory

state0 = [1,99]
N = np.sum(state0)
t = np.linspace(0,5,1000)
state = odeint(imperfectImmunity, state0, t)

plt.plot(t, state, linewidth = 5)
plt.plot(t, Rt(N,state[:,0], state[:,1]), linewidth =5)
plt.legend(('Infected', 'Susceptible', 'Recovered'))
```

```python
#Phase space
Ilim = [0,40]
Slim = [0,40]

h1 = np.linspace(Ilim[0], Ilim[1], 20)
h2 = np.linspace(Slim[0], Slim[1], 20)

H1, H2 = np.meshgrid(h1,h2)

u,v = np.zeros(H1.shape), np.zeros(H2.shape)

NJ, NK = H1.shape

for j in range(NJ):
    for k in range(NK):
        Istate = H1[j,k]
        Sstate = H2[j,k]
        
        statejk = [Istate, Sstate]
        
        ISdot = imperfectImmunity(statejk, [])
        u[j,k] = ISdot[0]
        v[j,k] = ISdot[1]

#Normalization
M = np.hypot(u,v)
M[M == 0] == 1.
u /= M
v /= M

plt.quiver(H1, H2, u,v, color = 'red')
plt.xlabel('Infected individuals', fontsize = 15)
plt.ylabel('Susceptible individuals', fontsize = 15)

plt.plot(state[:,0], state[:,1], linewidth =5)
plt.ylim((0,40))
plt.show()
```
