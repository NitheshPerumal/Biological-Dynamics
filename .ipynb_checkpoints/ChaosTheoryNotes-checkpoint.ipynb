# Chaos Theory 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint

%pylab inline
pylab.rcParams['figure.figsize'] = (12,10)
```

## Hallmarks of Chaotis Systems

* Determinism
* Boundedness (done blow up to infinity)
* Semi-irregular
* Highly sensitive to initital conditions
* Strange Attractors (picture tells a story)

```python
def lorenz(state, t0, sigma = 10, beta = 8./3, rho = 28.0):
    
    x = state[0]
    y = state[1]
    z = state[2]
    
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    
    return [dxdt, dydt, dzdt]
```

```python
N_trajectories = 2

#Vector of initial conditions (3 cols for each state var (x,y,z; requires its own IC) x N rows 
#where N is number of trajectories)
state0 = 10 + 0.01 * np.random.random((N_trajectories, 3))

t = np.linspace(0,100,1000)

state = np.asarray([odeint(lorenz, state0i, t)
                  for state0i in state0])

plt.plot(t, state[0,:, 0])
plt.plot(t, state[1,:, 0])

plt.legend(('Trajectory1', 'Trajectory2'), fontsize = 14)

plt.xlim(0,20)
```

```python
#What can we do to characterize a chaotic system?
#Quantitative method: Lyaponov Exponent

from scipy.spatial import distance

def compute_dst(state):
    dst = np.zeros((1, state.shape[1]))
    
    for i in np.arange(0, state.shape[1]):
        dst[0,i] = distance.euclidean(state[0,i,:], state[1,i,:])
    return(dst.T)

dst = compute_dst(state)

plt.plot(t,dst)
plt.xlabel('Time')
plt.ylabel('Distance between Trajectories')
plt.xlim(0,20)
```

```python
#Looking into universality of sensitivity on initial conditions
t = np.linspace(0,50,1000)
D = np.zeros((len(t), 100))

for i in np.arange(0,100):
    state0 = 10 + 0.01 * np.random.random((N_trajectories, 3))
    state = np.asarray([odeint(lorenz, state0i, t)
                  for state0i in state0])
    dst = compute_dst(state)
    
    D[:,i] = dst[:,0]
    
    plt.plot(dst)
```

```python
plt.semilogy(t, np.mean(D, axis = 1))

#Lyapunov exponent can be used to gain some qualitative insight into Chaotic system
D0 = mean(D[0,:])
lyapunov = 0.9 #approximated by guess and check
plt.semilogy(t, D0*np.exp(lyapunov*t))

plt.xlim(0,20)
plt.ylim(0,200)
plt.show()
```

## Holling-Tanner Model

* X = Plants
* Y = Herbivores
* Z = Carnivores

```python
def HollingTanner(state, t0, a1 = 5, b1 = 3, a2 = 0.1, b2 = 2, d1 = .4, d2 = .01 ):
    x = state[0]
    y = state[1]
    z = state[2]
    
    #Plants
    dxdt = x*(1-x) - ((a1*x)/(1 + b1*x)) * y
    
    #Herbivores
    dydt = (((a1*x)/(1 + b1*x)) * y) - d1*y - ((a2 * y)/(1+ b2 * y)) * z
    
    #Carnivores
    dzdt = (((a2 * y)/(1+ b2 * y)) * z) - d2 * z
    
    return [dxdt, dydt, dzdt]
```

```python
t = np.linspace(0,1000,10000)
state0 = np.array([1,.1,8])

state = odeint(HollingTanner, state0, t)

plt.plot(t, state)
plt.legend(('Plants' ,'Herbivores', 'Carniovores'))
```

```python
fig = plt.figure()
Ntimepoints = 1000
ax = fig.gca(projection = '3d')

plt.plot(state[0:Ntimepoints,0], state[0:Ntimepoints,1], state[0:Ntimepoints,2], color = 'r', linewidth =3)

ax.set_xlabel('Plants', fontsize = 20)
ax.set_ylabel('Herbivore', fontsize = 20)
ax.set_zlabel('Carnivore', fontsize = 20)
```

## Chaotic FitzHugh Nagumo Neuron Model

```python
def FHN(state, t):
    v = state[0]
    w = state[1]
    
    dv = -v*(v-a)*(v-1)-w + pulse(t, ton, toff, ion, ioff)#Na channel activation
    dw = eps*(v-gamma*w) #Recovery 
    
    return(dv,dw)

def pulse(t, ton, toff, ion, ioff):
    if (t%(ton+toff) < ton):
        u = ion
    elif (t%(ton+toff) >= ton):
        u = ioff
    return(u)

def computeU(t):
    U = np.zeros(len(t))
    for i in np.arange(0, len(t), 1):
        U[i] = pulse(t[i], ton, toff, ion, ioff)
    return(U)
```

```python
ton = 10
toff = 15
ion = 0.04
ioff = 0

time = np.linspace(0,1000,10000)

U = computeU(time)
plt.plot(U)
```

```python
eps = 0.01
a = 0.1
gamma = 2

state0 = [0,0]

state = odeint(FHN, state0, time)
```

```python
plt.plot(t, state[:,0])
plt.plot(t, computeU(time))
```

```python
fig = plt.figure()
Ntimepoints = 10000
ax = fig.gca(projection = '3d')

plt.plot(state[0:Ntimepoints,0], state[0:Ntimepoints,1], computeU(time), color = 'r', linewidth =3)
```
