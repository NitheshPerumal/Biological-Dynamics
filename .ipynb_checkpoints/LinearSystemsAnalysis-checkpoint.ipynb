# Linear Systems Analysis

## Topics
 * Eigen value and Eigenvector interpretation of dynamics
 * Plotting and investigating dynamic systems as function of time
 * Use Scipy's integrator to find numerical solutions to ODE's
 * Visualizing the set of dynamics a system can produce using Phase Portraits

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [12,12]
```

## The general system we analyze

# $\begin{bmatrix} \dot{x_1} \\ \dot{x_2} \end{bmatrix} = \begin{bmatrix} a_{1,1} & a_{1,2} \\ a{2,1} & a_2,2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$

```python
A = np.array([[1,3], [1,-1]])
evals, evecs = np.linalg.eig(A)

#Equivalences

print('trace(A) = ', np.trace(A), ' = sum of evals =', np.sum(evals))
print('det(A) = ', np.linalg.det(A),'= prodcut of evals = ', np.product(evals))

#Negtaive determinatn tells us that phase portrait will be saddlepoint
```

```python
#Define our diff eq in terms of function to use the integrator to solve it

def dxdt(state, t):
    
    #unpack state vars
    x1 = state[0]
    x2 = state[1]
    
    #Define the system
    dx1dt = A[0,0]*x1 + A[0,1]*x2
    dx2dt = A[1,0]*x1 + A[1,1]*x2
    
    #could also do dxdtvec = A @ state
    
    return(dx1dt, dx2dt)
```

```python
#Use integrator to find numerical solution
from scipy.integrate import odeint

#initital conditions
state0 = [-2.1, 2]

#vector of times to simulate for
t= np.linspace(0,2,1000)

#Run integration
state = odeint(dxdt, state0, t)
```

```python
#Plot the result
plt.plot(t,state, linewidth =3)
plt.legend(('x1', 'x2'), fontsize = 18)
plt.show()
```

```python
#Plot of phase plane of A
x1lim = [-2.6,5]
x2lim = [-2.6,2.6]

nPoints = 20
s1 = np.linspace(x1lim[0], x1lim[1], nPoints)
s2 = np.linspace(x2lim[0], x2lim[1], nPoints)

#Meshgrid - supplies set of matrices of coordinates to scan through phase plane
S1, S2 = np.meshgrid(s1,s2)

u, v = np.zeros(S1.shape), np.zeros(S2.shape)
NI, NJ = S1.shape

#Nested for loop to scan through phase space and calculate flow (dxdt) at each point
for i in range(NI):
    for j in range(NJ):
        
        x1 = S1[i,j]
        x2 = S2[i,j]
        
        #Create a state vector 
        X = np.array([[x1], [x2]])
        Xdot = A @ X  
        
        #Store values in the dxdt matricies u,v
        u[i,j] = Xdot[0]
        v[i,j] = Xdot[1]
```

```python
#Now plot the vectors
plt.quiver(S1, S2, u,v, color = 'red')
plt.xlabel('x1 values of state space', fontsize = 16)
plt.ylabel('x2 values of state space', fontsize = 16)
```

```python
#Normalize vectors so they are all the same size
M = (np.hypot(u,v)) #norm of the rate
M[M ==0] = 0.1

u /= M
v /= M

#Now plot the vectors
plt.quiver(S1, S2, u,v, color = 'red')
plt.xlabel('x1 values of state space', fontsize = 16)
plt.ylabel('x2 values of state space', fontsize = 16)
```

```python
# Plot eigenvector directions on top of the phase portraits
plt.plot(evecs[0,0]*s1, evecs[1,0]*s1, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'blueviolet')
plt.plot(evecs[0,1]*s2, evecs[1,1]*s2, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'violet')

plt.quiver(S1, S2, u,v, color = 'red')
plt.xlabel('x1 values of state space', fontsize = 16)
plt.ylabel('x2 values of state space', fontsize = 16)
```

```python
#Plot specific trajectory through the phase space
plt.plot(state[:,0], state[:,1], linewidth = 4)
plt.plot(evecs[0,0]*s1, evecs[1,0]*s1, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'blueviolet')
plt.plot(evecs[0,1]*s2, evecs[1,1]*s2, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'violet')

plt.quiver(S1, S2, u,v, color = 'red')
plt.xlabel('x1 values of state space', fontsize = 16)
plt.ylabel('x2 values of state space', fontsize = 16)
```

```python
t= np.linspace(0,2,1000)

state0 = [-2, 2]

#Run integration
state = odeint(dxdt, state0, t)

#Plot specific trajectory through the phase space
plt.plot(state[:,0], state[:,1], linewidth = 4)
plt.plot(evecs[0,0]*s1, evecs[1,0]*s1, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'blueviolet')
plt.plot(evecs[0,1]*s2, evecs[1,1]*s2, linewidth = 8, alpha = 0.6, linestyle = ':', color = 'violet')

plt.quiver(S1, S2, u,v, color = 'red')
plt.xlabel('x1 values of state space', fontsize = 16)
plt.ylabel('x2 values of state space', fontsize = 16)
```
